```python
import pandas as pd
data = pd.read_csv("https://data.wprdc.org/dataset/c6265ffb-c86e-4a2c-870b-1aaa0772c449/resource/14abb000-d066-4728-8ffa-227883d6c331/download/wage-or-salary-income-in-the-past-12-months-for-households.csv",
                  index_col='Id')
data.head(10)
```

So I have data about how many people are living in a neighborhood "With wage or salary income" and with "No wage or salary income".
My metric is the percentage of the population in a neighborhood that has an income.
I'm going to make a dataframe that holds values equivalent to the percentage of people with wage/salary income out of the whole population of that area.

```python
percs = [0.0]*len(data.index)
i = 0
for item in data.iterrows():
    try:
        percent = item[1][3]/item[1][1]
    except:
        percent = 0.0
    percs[i]=percent
    i+=1
```

So now I have an array with all the percentages (Total people with income)/(Total people in neighborhood)
Let's add it to the dataframe in a column called 'Percent with Income'

```python
data['Percent with Income'] = percs
data.head(10)
```

Now let's find the top 10 Neighborhoods in terms of income

```python
data.sort_values('Percent with Income', ascending = False).head(10)
```
