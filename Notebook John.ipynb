```python
import pandas as pd
data = pd.read_csv("https://data.wprdc.org/dataset/c6265ffb-c86e-4a2c-870b-1aaa0772c449/resource/14abb000-d066-4728-8ffa-227883d6c331/download/wage-or-salary-income-in-the-past-12-months-for-households.csv",
                  index_col='Id')
data.head(10)
```

So I have data about how many people are living in a neighborhood "With wage or salary income" and with "No wage or salary income".
I want to find the percentage of the population in a neighborhood that has an income.
I'm going to make a dataframe that holds values equivalent to the percentage of people with wage/salary income out of the whole population of that area. But first I'll start with calculating those percentages

```python
percs = [0.0]*len(data.index)
i = 0
#iterate through the rows
for item in data.iterrows():
    try:#attempt to do the division
        percent = item[1][3]/item[1][1]
    except:#any errors will result in an assumed 0%
        percent = 0.0
    percs[i]=percent
    i+=1
```

So now I have an array with all the percentages (Total people with income)/(Total people in neighborhood)
Let's add it to the dataframe in a column called 'Percent with Income'

```python
data['Percent with Income'] = percs
data.head(10)
```

Now let's find the top 10 Neighborhoods in terms of income

```python
data.sort_values('Percent with Income', ascending = False).head(10)
```

So the neighborhoods that have the most people making money are:
1. South Shore
2. Fairywood
3. North Shore
4. Friendship
5. Esplen
6. Mt. Oliver
7. Strip District
8. Central Oakland
9. Ridgemont
10. Summer Hill

But if you notice, South shore only has 8 people, which has no real impact. So I'll have to account for margin of error. I'll subtract the margin of error from the totals (specifically for those with wage or salary income) and then divide it by the estimate total. This should make things a bit more accurate.

```python
data['Adjusted Percent with Income']=data['Estimate; Total: - With wage or salary income'].sub(data['Margin of Error; Total: - With wage or salary income']).div(data['Estimate; Total:'])
data.sort_values('Adjusted Percent with Income', ascending = False).head(10)
```

So by my new metric of an adjusted percentage of the population with income, the Top Neighborhood in Pittsburgh is
# Fairywood
